##Linear Regression to predict score of student given number of hours studied.

```python
#importing libraries required
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
#Read data from link
url="http://bit.ly/w-data"
r=pd.read_csv(url)
print("Data import successful")
r.head(15)
```

```python
#Plotting the data
r.plot(x='Hours',y='Scores',style='x',color='red')
plt.title("Hours vs Percentage Score")
plt.xlabel("Hours studied")
plt.ylabel("Percentage Score")
plt.show
```

#**Preparing Data**

```python
#Divide data into attributes(inputs)and labels(outputs)
X=r.iloc[:,:-1].values #gets values of first col; .values removes axes labels
Y=r.iloc[:,1].values #gets values of 2nd col
```

```python
#Splitting data into training and testing sets
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test= train_test_split(X,Y,random_state=0,test_size=0.2)
```

#**Training**

```python
from sklearn.linear_model import LinearRegression
regressor= LinearRegression()
regressor.fit(X_train,Y_train)
print("Training Complete")
```

```python
#Plot regression line
line= regressor.coef_*X+regressor.intercept_
#Plot for test data
plt.scatter(X,Y)
plt.plot(X,line)
plt.show()
```

#**Make Predictions**

```python
y_pred = regressor.predict(X_test) # Predicting the scores
```

```python
#Compare actual with predicted
df=pd.DataFrame({'Actual':Y_test,'Predicted':y_pred})
df
```

#**Evaluating Model**

```python
from sklearn import metrics
#MeanSquareError
print('Mean Absolute Error:',
      metrics.mean_absolute_error(Y_test, y_pred))
```
